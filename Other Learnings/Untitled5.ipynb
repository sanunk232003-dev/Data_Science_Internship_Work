```python
import math
from abc import ABC,abstractmethod
class Shape(ABC):
  @abstractmethod
  def area(self):
    pass
class Circle(Shape):
  def area(self,r):
    return math.pi*r**2
class Rectangle(Shape):
  def area(self,length,width):
    return length*width
c=Circle()
print(c.area(5))


d=Rectangle()
print(d.area(2,3))





```
