```python
import pandas as pd
data=pd.read_csv("/content/Student_Performance_Teaching_Effectiveness.csv")
data.head()
```

```python
x=data[['Study_Hours','Attendance','Assignments']]
y=data['Final_Grade']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
predictions=model.predict(x_test)
print(predictions)
```

```python
print("Actual values",list(y_test))
print("Predicted values",predictions)
```

```python
import matplotlib.pyplot as plt
plt.bar(data['Attendance'],data['Final_Grade'])
plt.xlabel('Attendance')
plt.ylabel('Final_Grade')
plt.title('Attendance vs Final_Grade')
plt.show()
```
