```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv("/content/test.csv")
#data.head()
display(data)
```

```python
x=data['Age']
y=data['Salary']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train.values.reshape(-1, 1), y_train)
```

```python
predictions=model.predict(x_test.values.reshape(-1, 1))
print(predictions)
```

```python
import matplotlib.pyplot as plt
plt.scatter(data['Age'],data['Salary'])
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Age vs Salary')
plt.show()
```

```python
print("Actual values",list(y_test))
print("Predicted values",predictions)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
model_dt=DecisionTreeRegressor()
model_dt.fit(x_train.values.reshape(-1, 1),y_train)
pred_dt=model_dt.predict(x_test.values.reshape(-1, 1))
print("DecisionTree Predictions",pred_dt)
print("Decision Tree MAE:",mean_absolute_error(y_test,pred_dt))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
model_rf=RandomForestRegressor()
model_rf.fit(x_train.values.reshape(-1, 1),y_train)
pred_rf=model_rf.predict(x_test.values.reshape(-1, 1))
print("Random Forest Predictions",pred_rf)
print("Random Forest MAE:",mean_absolute_error(y_test,pred_rf))
```

```python
from sklearn.svm import SVR
model_svr=SVR()
model_svr.fit(x_train.values.reshape(-1, 1),y_train)
pred_svr=model_svr.predict(x_test.values.reshape(-1, 1))
print("SVR Predictions",pred_svr)
print("SVR MAE:",mean_absolute_error(y_test,pred_svr))
```

```python
from sklearn.neighbors import KNeighborsRegressor
model_knn=KNeighborsRegressor(n_neighbors=3)
model_knn.fit(x_train.values.reshape(-1, 1),y_train)
pred_knn=model_knn.predict(x_test.values.reshape(-1, 1))
print("KNN Predictions",pred_knn)
print("KNN MAE:",mean_absolute_error(y_test,pred_knn))
```

```python
print("Linear Regression Error",mean_absolute_error(y_test,predictions))
print("Decision Tree Error",mean_absolute_error(y_test,pred_dt))
print("Random Forest Error",mean_absolute_error(y_test,pred_rf))
print("SVR Error",mean_absolute_error(y_test,pred_svr))
print("KNN Error",mean_absolute_error(y_test,pred_knn))
```
