```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
    def s1(self):
        print(f"I am good in physics.I got {self.marks}")
class ma(Student):
  def __init__(self,name,marks):
    super().__init__(name,marks)
  def cat(self):
      print(f" I got {self.marks} in physics")


s=ma("John", 80)
print(s.name)
#print(s.__marks)
s.cat()
s.s1()
```

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self._age = age
        self.__grade = "A"


    def show_info(self):
        print(f"Name: {self.name}, Age: {self._age}")
    def _update_age(self, new_age):
        self._age = new_age
    def __calculate_grade(self):
        return f"Grade: {self.__grade}"
    def get_grade(self):
        return self.__calculate_grade()



student = Student("Alice", 20)
student.show_info()
student._update_age(21)
print(student.get_grade())

```

```python
class Employee:
    def __init__(self, name, employee_id, salary):
        self.name = name
        self.__salary = salary
        print(f"my name is {name} my id is {employee_id} my salary is{salary}")

    def show_salary(self):
        return self.__salary

    def access_inside_class(self):
        print("Inside Employee class:")
        print(self.name)
        print(self._employee_id)
        print(self.__salary)
df=Employee("rajesh",101,"12000")

df.show_salary()
df.access_inside_class()
```

```python
class Manager(Employee):
    def access_inside_subclass(self):
        print("Inside Manager (Subclass):")
        print(self.name)
        print(self._employee_id)

        # print(self.__salary)
        print(self._Employee__salary)
```

```python
class Developer(Manager):
    def access_inside_developer(self):
        print("Inside Developer (Grandchild class):")
        print(self.name)             # ✅ Public
        print(self._employee_id)     # ⚠️ Protected
        # print(self.__salary)       # ❌ Still not accessible
```
