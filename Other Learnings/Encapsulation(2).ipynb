```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.__marks = marks
    def s1(self):
        print(f"i am good in physics {self.__marks}")
class ma(Student):
  def __init__(self,name,marks):
    super().__init__(name,marks)
  def cat(self):
      print(f" I got{self.__marks}in physics")


s=ma("John", 80)
print(s.name)
#print(s.__marks)
s.cat()
s.s1()
```

```python
class Student:
    def __init__(self, name, age):
        self.name = name            # Public attribute
        self._age = age             # Protected attribute (convention)
        self.__grade = "A"          # Private attribute (name mangling)

    # Public method
    def show_info(self):
        print(f"Name: {self.name}, Age: {self._age}")

    # Protected method
    def _update_age(self, new_age):
        self._age = new_age

    # Private method
    def __calculate_grade(self):
        return f"Grade: {self.__grade}"

    # Public method accessing private
    def get_grade(self):
        return self.__calculate_grade()


# Usage
student = Student("Alice", 20)
student.show_info()          # ✅ Public access
student._update_age(21)      # ⚠️ Possible, but discouraged
print(student.get_grade())   # ✅ Access private via public method

# Direct private access will fail:
# print(student.__grade)     # ❌ AttributeError
```

```python
class Employee:
    def __init__(self, name, employee_id, salary):
        self.name = name              # PUBLIC
        self._employee_id = employee_id  # PROTECTED
        self.__salary = salary
        print(f"my name is {name} my id is {employee_id} my salary is{salary}")     # PRIVATE

    def show_salary(self):  # PUBLIC method
        return self.__salary

    def access_inside_class(self):
        print("Inside Employee class:")
        print(self.name)          # ✅ Public
        print(self._employee_id)  # ✅ Protected
        print(self.__salary)      # ✅ Private
df=Employee("rajesh",101,"12000")

df.show_salary()
df.access_inside_class()
```

```python
class Manager(Employee):
    def access_inside_subclass(self):
        print("Inside Manager (Subclass):")
        print(self.name)             # ✅ Public
        print(self._employee_id)     # ⚠️ Protected (allowed by convention)

        # print(self.__salary)       # ❌ ERROR (private)
        print(self._Employee__salary)  # ⚠️ Name mangling (not recommended)
```

```python
class Developer(Manager):
    def access_inside_developer(self):
        print("Inside Developer (Grandchild class):")
        print(self.name)             # ✅ Public
        print(self._employee_id)     # ⚠️ Protected
        # print(self.__salary)       # ❌ Still not accessible
```

```python
emp = Employee("John", "E101", 50000)

print("From external code:")
print(emp.name)             # ✅ Public
print(emp._employee_id)     # ⚠️ Technically accessible (but DON'T)

# print(emp.__salary)       # ❌ ERROR
print(emp._Employee__salary)  # ⚠️ Name mangling (bad practice)
```

```python
class BankAccount:
  def __init__(self,account_holder,balance,pin):
    self.account_holder=account_holder
    self._balance=balance
    self.__pin=pin
  def display_account_info(self):
    print("Account holder name=",self.account_holder,"Balance=",self._balance)
  def deposit(self,amount):
    if amount>0:
      self._balance+=amount
      print("The balance is",self._balance)
    else:
      print("The amount not enough")
  def validate_pin(self,pin):
    if pin==self.__pin:
      print("The pin is valid")
    else:
      print("The pin is invalid")
  def withdraw(self,amount):
      if amount>0:
          self._balance-=amount
          print("The amount after withdraw is", self._balance)
      else:
          print("Invalid amount")



account=BankAccount(account_holder="Sahana",balance=10000,pin="99729")
account.display_account_info()
account.deposit(10)
account.validate_pin("99729")
account.withdraw(450)
```
