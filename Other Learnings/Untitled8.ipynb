```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow

# create a white image
img = np.ones((300, 300, 3), dtype=np.uint8) * 255

# draw a rectangle
cv2.rectangle(img, (50, 50), (250, 250), (0, 0, 255), 5)

cv2_imshow(img)
```
