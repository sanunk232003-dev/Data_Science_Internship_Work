```python
from abc import ABC,abstractmethod
class Animal(ABC):
  @abstractmethod
  def sound(self):
    pass
class Dog(Animal):
  def sound(self):
    print("Bark Bark")
class Cat(Animal):
  def breed(self):
    print("the breed is german")
  def sound(self):
    print("Meow Meow")
animals=[Dog(),Cat()]
for a in animals:
  a.sound()




```
