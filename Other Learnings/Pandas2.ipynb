```python
import pandas as pd

df = pd.DataFrame({
'col1': [1, 2, 3, 4],
'col2': [5, 6, 7, 8],
'col3': [9, 10, 11, 12]
}, index=['a', 'b', 'c', 'd'])
print("Original DataFrame:")
print(df)
# Swap rows 'a' and 'c' using loc
df.loc[['a', 'c'], :] = df.loc[['c', 'a'], :].values
print("\nDataFrame after swapping rows 'a' and 'c' using loc:")
print(df)
# Swap columns 'col1' and 'col3' using iloc
df.iloc[:, [0, 2]] = df.iloc[:, [2, 0]].values
print("\nDataFrame after swapping columns 'col1' and 'col3' using iloc:")
```
