```python
# Program 1: Handling division by zero
try:
    num = int(input("Enter a number: "))
    result = 10 / num
    print("Result:", result)
except ZeroDivisionError:
    print("Error: Cannot divide by zero!")
```

```python
# Program 2: Handling multiple error types
try:
    num = int(input("Enter a number: "))
    result = 10 / num
    print("Result:", result)
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
except ValueError:
    print("Error: Please enter a valid integer.")
```

```python
# Program 3: Adding finally block
try:
    num = int(input("Enter a number: "))
    result = 10 / num
    print("Result:", result)
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
except ValueError:
    print("Error: Please enter a valid integer.")
finally:
    print("Execution finished. Cleaning up resources...")
```

```python
# Program 4: File handling with cleanup
try:
    f = open("/content/thagu.txt", "r")
    content = f.read()
    print("File content:", content)
except FileNotFoundError:
    print("Error: File not found.")
finally:
    print("Closing file (if opened).")
    try:
        f.close()
    except NameError:
        print("File was never opened.")
```

```python
# Program 5: Using else block
try:
    num = int(input("Enter a number: "))
    result = 10 / num
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
except ValueError:
    print("Error: Please enter a valid integer.")
else:
    print("Division successful! Result:", result)
finally:
    print("Program execution complete.")
```
