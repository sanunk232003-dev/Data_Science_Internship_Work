```python
class Student:
  def __init__(self,name,marks):
    self.name=name
    self.marks=marks
  def s1(self):
    print(f"i am good in physics {self.marks}")
class ma(Student):
  def __init__(self,name,marks):
    super().__init__(name,marks)
  def cat(self):
    print(f" I got {self.marks} in Physics")
s=ma("John",80)
print(s.name)
s.cat()
s.s1()
```

```python
class Animal:
  def __init__(self,name,breed):
    self.name=name
    self.__breed=breed
  def s1(self):
    print("the animal name is", self.name ,"the breed is", self.__breed)
c=Animal("cat","persian")
c.s1()
```
