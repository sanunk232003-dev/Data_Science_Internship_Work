```python
import pandas as pd
data=pd.read_csv("/content/ml2.csv")
data.head()
```

```python
x=data[['Maths','Kannada','English','ComputerScience']]
y=data['Total_score']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
predictions=model.predict(x_test)
print(predictions)
```

```python
print("Actual values",list(y_test))
print("Predicted values",predictions)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
model_dt=DecisionTreeRegressor()
model_dt.fit(x_train,y_train)
pred_dt=model_dt.predict(x_test)
print("DecisionTree Predictions",pred_dt)
print("Decision Tree MAE:",mean_absolute_error(y_test,pred_dt))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
model_rf=RandomForestRegressor()
model_rf.fit(x_train,y_train)
pred_rf=model_rf.predict(x_test)
print("Random Forest Predictions",pred_rf)
print("Random Forest MAE:",mean_absolute_error(y_test,pred_rf))
```

```python
from sklearn.svm import SVR
model_svr=SVR()
model_svr.fit(x_train,y_train)
pred_svr=model_svr.predict(x_test)
print("SVR Predictions",pred_svr)
print("SVR MAE:",mean_absolute_error(y_test,pred_svr))
```

```python
from sklearn.neighbors import KNeighborsRegressor
model_knn=KNeighborsRegressor(n_neighbors=3)
model_knn.fit(x_train,y_train)
pred_knn=model_knn.predict(x_test)
print("KNN Predictions",pred_knn)
print("KNN MAE:",mean_absolute_error(y_test,pred_knn))
```

```python
print("Linear Regression Error",mean_absolute_error(y_test,predictions))
print("Decision Tree Error",mean_absolute_error(y_test,pred_dt))
print("Random Forest Error",mean_absolute_error(y_test,pred_rf))
print("SVR Error",mean_absolute_error(y_test,pred_svr))
print("KNN Error",mean_absolute_error(y_test,pred_knn))
```

```python
import matplotlib.pyplot as plt

subjects = ['Maths','Kannada','English','ComputerScience']

for subject in subjects:
    plt.scatter(data[subject], data['Total_score'], label=subject)

plt.ylabel('Total Score')
plt.xlabel('Subject Score')
plt.title('Subject Scores vs Total Score')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

subjects = ['Maths','Kannada','English','ComputerScience']
mean_scores = data[subjects].mean()
plt.figure(figsize=(10, 6))
plt.bar(mean_scores.index, mean_scores.values, color='skyblue')
plt.xlabel('Subjects')
plt.ylabel('Average Score')
plt.title('Average Scores per Subject')
plt.ylim(0, 50)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
