```python
# @title
import pandas as pd
df = pd.read_csv("StudentsPerformance.csv")
df.loc[8, "math score"] = 64

df["test preparation course"].fillna("none", inplace=True)


avg_reading = df["reading score"].mean()
df["reading score"].fillna(avg_reading, inplace=True)


math_min = df["math score"].min()
math_max = df["math score"].max()
math_mean = df["math score"].mean()
math_median = df["math score"].median()
math_mode = df["math score"].mode()[0]

print("Math Score Stats:")
print("Min: {math_min}")
print("Max: {math_max}")
print("Mean: {math_mean:.2f}")
print("Median: {math_median}")
print("Mode: {math_mode}")
```
