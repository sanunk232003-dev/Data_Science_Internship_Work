```python
series1 = pd.Series([1, 2, 3])
series2 = pd.Series([4, 5, 6])
print(type(series1))
print("Before swap:")
print("series1:\n",series1)
print("series2:", series2)
series1, series2 = series2, series1
print("\nAfter swap:")
print("series1:\n", series1)
print("series2:", series2)
```
