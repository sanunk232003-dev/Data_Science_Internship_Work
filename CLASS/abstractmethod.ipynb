```python
from abc import ABC,abstractmethod
class Appliance(ABC):
  @abstractmethod
  def turn_on(self):
    pass
class WashingMachine(Appliance):
  def turn_on(self):
    print("WashingMachine turned on")
class Microwave(Appliance):
  def turn_on(self):
    print("Microwave turned on")
Appliance=[WashingMachine(),Microwave()]
wm=WashingMachine()
wm.turn_on()
m=Microwave()
m.turn_on()





```
